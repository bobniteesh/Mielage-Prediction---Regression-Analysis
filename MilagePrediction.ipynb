**Milage Prediction**

**Source:**
This dataset was taken from the Statl_ib library which is maintained at Carnegie Mellon University. The dataset was used in the 1983 American
Statistical Association Exposition.
**Data Set Information:**
This dataset is a slightly modified version of the dataset provided in the StatLib library, In line with the use by Ross Quinlan (1993) in predicting
the attribute •rnpg•, 8 of the original instances were removed because they had unknown values for the ernpg• attribute. The original dataset is
available in the file
'The data concerns city-cycle fuel consurnption in miles per gallon, to be predicted in terms of 3 multivalued discrete and 5 continuous
attributes.- (Quinlan, 1993)

**Attribute Information:**
1. mpg:
2. cylinders: rnulti-valued discrete
3. displacement: continuous
4. horsepower. continuous
5. weight: continuous
6. acceleration: continuous
7. model year: multi-valued discrete
8. origin: multi-valued discrete
9. car name: String (unique for each instance)

**Data Preprocessing**

**Remove Missing Values**

```python
df=df.dropna()
```

```python
df.info()
```

**Data Visualization**

```python
sns.pairplot(df,x_vars =[ 'displacement','horsepower','weight','acceleration','mpg'],y_vars=['mpg']);
```

**Define Target Variable y and Feature X**

```python
df.columns
```

```python
y=df['mpg']
```

```python
y.shape
```

```python
X =df[['displacement','horsepower','weight','acceleration']]
```

```python
X.shape
```

```python
X
```

**Scaling Data**

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X = ss.fit_transform(X)
```

```python
X
```

```python
pd.DataFrame(X).describe()
```

After Standardization Mean is Zero and Standard Deviation is One

**Train Test Split Data**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, random_state =2529)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

**Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

Mileage 23.4 - 1.05Displacemet - 1.68Horsepower - 4.1Weight - 0.115Acceleration + error

**Predict Test Data**

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

**Model Accuracy**

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```
